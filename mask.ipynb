```python
import cv2
```

```python
face_model = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
nose_model =cv2.CascadeClassifier("haarcascade_mcs_mouth.xml")
mouth_model = cv2.CascadeClassifier("haarcascade_mcs_nose.xml")
```

```python
video = cv2.VideoCapture(0)
while True :
    suc,img = video.read()
    img = cv2.flip(img, 1)
    grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_model.detectMultiScale(grey_img,minNeighbors = 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,225),3)
        face_region = grey_img[y:y+h, x:x+w]
        nose = nose_model.detectMultiScale(face_region,scaleFactor=1.3,minNeighbors = 5)
        mouth = mouth_model.detectMultiScale(face_region,scaleFactor=1.25,minNeighbors = 5)

        if len(nose) == 0 and len(mouth) == 0 :
            cv2.putText(img,"Masked Person",(x,y-5),2,.5,(0,0,225),2)
        else :
            cv2.putText(img,"UnMasked Person",(x,y-5),2,.5,(0,0,225),2)

       
    cv2.imshow("web Cam",img)
    if cv2.waitKey(1) & 0XFF == ord("q"):
        break

video.release()
cv2.destroyAllWindows()

        
```
